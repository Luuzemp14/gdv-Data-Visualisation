### Libraries einlesen

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Datensatz "Pokémon" einlesen

```python
data = pd.read_csv("Pokemon Datensatz.csv")

#Kurzer Überblick

data.head(8)
```

### Überblick über die Daten verschaffen

```python
data.info()
```

### Spaltennamen ändern, damit es keine Abstände dazwischen gibt

```python
pokemon = data.rename(columns = {"Type 1":"Type_1", "Type 2":"Type_2","Sp. Atk":"Sp.atk","Sp. Def":"Sp.def"})
```

### Erste Plots erstellen

```python
# Zeige ob es mehr Feuer oder Wasser Pokemon gibt
fire_pokemon = pokemon[pokemon.Type_1 == "Fire"]
water_pokemon = pokemon[pokemon.Type_2 == "Water"]
```

```python
pokemon_type = pokemon["Type_1"]
pokemon_hp = pokemon["HP"]
plt.scatter(pokemon_type, pokemon_hp)
plt.title("Typen und ihre HP Werte")
plt.show()
```

```python
pokemon_hp = pokemon.HP
pokemon_speed = pokemon.Speed
plt.scatter(pokemon_hp, pokemon_speed)
plt.xlabel("HP")
plt.ylabel("Speed")
plt.show()
```
